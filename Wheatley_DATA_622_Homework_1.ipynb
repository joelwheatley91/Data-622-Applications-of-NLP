3. Extract Headings from a Wikipedia Page

Fetch the Wikipedia page for Natural Language Processing.

Extract and print all the headings (\<h1>, \<h2>, \<h3>) from the page.

```python
url = "https://en.wikipedia.org/wiki/Natural_language_processing"


#Wikipedia was blocking my scraping, so I set up an alias User-agent to mimic normal web browser.
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

#Pull and Parse NLP wiki page
r_wiki = requests.get(url, headers = headers)

html_wiki = r_wiki.text

soup_wiki = BeautifulSoup(html_wiki, 'html.parser')

#Print out all lines labled h1, h2, h3
print("Header 1")
for header1 in soup_wiki.find_all('h1'):
  print(header1.get_text())
print()
print("Header 2")
for header2 in soup_wiki.find_all('h2'):
  print(header2.get_text())

print()
print("Header 3")
for header3 in soup_wiki.find_all('h3'):
  print(header3.get_text())
```

4. Extract Links from a Page

From the Wikipedia page above, extract and print all the URLs (href values) of links found
on the page.

```python
#href is label for hyperlinks, so anything that is one is what we are looking for.
for link in soup_wiki.find_all("a", 'href'==True):
  print(link.get_text())
```

5. Extract and Save Text

Extract the first paragraph (\<p>) from the Wikipedia page and save it to a local text file called nlp_intro.txt.

```python
#Get the first paragrah (item 0 in the array of paragraphs)
paragraph1 = soup_wiki.find_all("p")[0].get_text()

#Write it to the file designated
with open("nlp_intro.txt", "w") as f:
  f.write(paragraph1)
```
